## 3. Data Cleaning - Additional

#### Separating Columns:

```python
df.head()
```

```python
df = df.drop(columns = 'Unnamed: 0')
```

#### Seperating the Boat type column:

```python
#checking for values in boat type column
df['Boat Type'].value_counts(dropna = False)
```

```python
# creating new columns 'Boat Class' and 'Boat Subclass'
df[['Boat Class','Boat Subclass']]= df['Boat Type'].str.split(',', 1, expand=True)
df[['Boat Subclass','Other']]= df['Boat Subclass'].str.split(',', 1, expand=True)
df['Boat Subclass']=df['Boat Subclass'].str.replace('/','',regex=True)
df = df.drop(['Boat Type', 'Other'],axis=1)
```

```python
df.head(15)
```

```python
#checking values in boat class column
df['Boat Class'].value_counts(dropna = False)
```

```python
#checking values in boat subclass column
df['Boat Subclass'].value_counts(dropna = False)
```

There isn't enough data to make this column worth keeping, upon review the boat subclass will be removed

```python
#dropping boat subclass column
df.drop(columns=['Boat Subclass'], inplace=True)
```

```python
df.columns
```

#### Seperating the Boat Condition Column:

```python
# creating new columns 'Boat Status' and 'Boat Fuel Type'
df[['Boat Status','Boat Fuel Type']]= df['Boat Condition'].str.split(',', 1, expand=True)
df = df.drop(['Boat Condition'],axis=1)
```

```python
df.head(15)
```

```python
#checking values in boat status column
df['Boat Status'].value_counts(dropna = False)
```

Updating the diesel, unleaded, and electric rows to be 'Not Disclosed'

```python
# updating value in boat status column
df['Boat Status'] = df['Boat Status'].replace(['Diesel','Unleaded','Electric'],['Not Disclosed','Not Disclosed','Not Disclosed'])
```

```python
#checking values in boat status column
df['Boat Status'].value_counts(dropna = False)
```

```python
#checking values in boat fuel type column
df['Boat Fuel Type'].value_counts(dropna = False)
```

Updating 'none' valued to 'Not Disclosed'

```python
# updating value in boat fuel type column
df['Boat Fuel Type'] = df['Boat Fuel Type'].fillna('Not Disclosed')
```

```python
#checking values in boat fuel type column
df['Boat Fuel Type'].value_counts(dropna = False)
```

```python
df.head(15)
```

### Reviewing Remaining Columns

```python
# Check for missing values
df.isnull().sum()
```

```python
#dropping null city values
df=df.dropna()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
df.shape
```

```python
# Checking for mixed type data
for col in df.columns.tolist():
      weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
      if len (df[weird]) > 0:
        print (col)
```

no columns with mixed data

```python
# updating country column to remove any rows with numeric variables
df['Country'] = df['Country'].apply(lambda x: pd.to_numeric(x, errors = 'ignore'))
```

```python
# re-checking values in country column
df['Country'].value_counts(dropna = False)
```
